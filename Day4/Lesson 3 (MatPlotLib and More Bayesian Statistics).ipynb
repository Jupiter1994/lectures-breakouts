# The basic plot...

Here we create a simple two dimensional plot.

```python
import numpy as np #matplotlib can deal well with numpy arrays

x = np.linspace(0,10,100)
y = x**2

plt.plot(x,y)
plt.show()
```

All good plots should have a two critical things...

1) X and Y labels with units.

2) A legend to show what each curve is.

There are two optional things which I find generally useful...

3) A top title

4) Grid lines for easily reading off datapoints

```python
plt.plot(x,y,label="Car 1")
plt.xlabel("time (sec)")
plt.ylabel("velocity (km/s)")
plt.legend()
plt.show()
```

```python
plt.plot(x,y,label="Car 1")
plt.xlabel("time (sec)")
plt.ylabel("velocity (km/s)")
plt.title("Acceleration of a Car")
plt.legend()
plt.grid()
plt.show()
```

```python
# Example from Nick Kern

# generate x-data
time = np.linspace(0, 1, 50)

# generate y-data
freq1 = 2 * np.pi * 1   # 2pi
freq2 = 2 * np.pi * 2   # 4pi
freq3 = 2 * np.pi * 3   # 6pi

y1 = np.sin(freq1*time)
y2 = np.sin(freq2*time)
y3 = np.sin(freq3*time)
```

```python
fig = plt.figure(figsize=(12,6))

# make a single subplot
ax = fig.add_subplot(1, 1, 1)

# create background grid
ax.grid(True)

# plot cosine as lines, no points, collect output of function
# note the data aren't continuous, it's just a plotting illusion
p1 = ax.plot(time, y1, color='green')
p2 = ax.plot(time, y2, color='blue')
p3 = ax.plot(time, y3, color='red')

# set x and y limits
ax.set_xlim(-0.1, 1.1)
ax.set_ylim(-1.1, 1.1)

# set x and y labels
ax.set_xlabel('time [sec]', fontsize=16)
ax.set_ylabel('sine function', fontsize=16)

# set title
ax.set_title('Plotting sine function with varying frequencies', fontsize=16)

# create legend
ax.legend([p1[0], p2[0], p3[0]], ['$2\pi$', '$4\pi$', '$6\pi$'], fontsize=16) #Note you can use LaTeX as labels
```

## More Bayesian Statistics

```python
# Example from Friday
import math

def binom(x,N,Theta):
    return math.factorial(N)/(math.factorial(x)*math.factorial(N-x))*np.power(Theta,x)*np.power(1-Theta,N-x)

theta = np.linspace(0, 1, 110)
prior = np.ones(len(theta))/len(theta)
lik = binom(4,5,theta) # Binomail Distribution

h = prior*lik
post = h/np.sum(h)
plt.plot(theta,post)
```

Lets imagine the true value is 0.8... What happens when we collect more data? Lets say we pick 5 more avocados and 3 of them are ripe... How do we updated our posterior?


```python
prior = post
lik = binom(3,5,theta) # Binomail Distribution

h = prior*lik
post_update = h/np.sum(h)

plt.plot(theta,post,"b:")
plt.plot(theta,post_update)

prior = np.ones(len(theta))/len(theta)

lik = binom(7,10,theta) # Binomail Distribution

h = prior*lik
post_together = h/np.sum(h)
plt.plot(theta,post_together,"r--")

```

There is nothing "special" about a prior vs. likelihood! In the end after multiple experiments we could take them in turn, iteratively updating the posterior and using it as a prior for each one, or we could use them all together!

## Summarizing the Posterior

Using there whole distribution is generally unwieldy... We often want a more concise summary statistics. For Gaussian distributions this fairly easy as we have a symmetric distribution that can summarized by a mean and a standard deviation. Often in real life posteriors end up looking fairly gaussian! In the above example, in the large N limit, we actually get asymptotically a gaussian.

```python

N = 10
theta = np.linspace(0, 1, 110)
prior = np.ones(len(theta))/len(theta)
lik = binom(4*N,5*N,theta) 
h = prior*lik
post = h/np.sum(h)
plt.plot(theta,post)
```

This is a result of the "Central Limit Theorem"!

CLT is a statistical theory that states that given a sufficiently large sample size from a population with a finite level of variance and sufficient mutual independence, the mean of all samples from the same population will be approximately equal to the mean of the population.

This is why gaussiansm, particularly multivariant gaussians, are very important to understand as they appear almost literally everywhere!

```python
data = np.random.normal(size=[1000,2])
data_new = np.array([data[:,0],(data[:,1]+data[:,0])/sqrt(2)]).transpose()


plt.plot(data_new[:,0],data_new[:,1],"r.")
plt.xlabel("x")
plt.ylabel("y")
```

```python
figure = corner.corner(data_new,quantiles=[0.16, 0.5, 0.84],smooth=1,show_titles=True)
```

In this example, we would say that variables x and y "covary" or have a covariance.
