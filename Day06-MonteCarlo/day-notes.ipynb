```python
def circle(x): #x^2 + y^1 = 1
    return np.sqrt(1-x**2) # ^ not the power sign
```

```python
x_array = np.linspace(0,1,100)

plt.plot(x_array,circle(x_array))
```

```python
n_inside = 0
n_outside = 0
for i in range(0,1000):
    x_point = np.random.random()
    y_point = np.random.random()

    if circle(x_point)>y_point: #in the circle
        plt.plot(x_point,y_point,"ro")
        n_inside += 1 
    else:
        plt.plot(x_point,y_point,"bo")
        n_outside += 1 
    
print(n_inside/(n_inside+n_outside), np.pi/4)
plt.plot(x_array,circle(x_array))
    
```

```python
def slow_pi():
    n_inside = 0
    n_outside = 0
    for i in range(0,1000000): #FOR LOOP!
        x_point = np.random.random()
        y_point = np.random.random()

        if circle(x_point)>y_point: #in the circle
           # plt.plot(x_point,y_point,"ro")
            n_inside += 1 
        else:
           # plt.plot(x_point,y_point,"bo")
            n_outside += 1 

    print(n_inside/(n_inside+n_outside), np.pi/4)
#plt.plot(x_array,circle(x_array))
    
```

```python
def numpy_pi():
    n = 10000000

    x_points = np.random.random(n)
    y_points = np.random.random(n)

    n_inside = np.sum(circle(x_points)>y_points)

    print(n_inside/(n), np.pi/4)
```
